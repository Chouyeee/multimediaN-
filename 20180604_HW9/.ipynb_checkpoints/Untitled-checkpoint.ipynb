```python
from skimage import io
import matplotlib.pyplot as plt
import math

def posterize(image):
    rows,cols,dims  = image.shape
    for i in range(0,rows):
           for j in range(0,cols):
                for k in range(0,3):
                    color = image[i,j,k]
                    image[i,j,k] = math.floor(color/64)*64+31
    return image

img = io.imread('Santa Claus.PNG')
io.imshow(img)

img2 = posterize(img)
plt.figure()
io.imshow(img2)
io.show()
```

```python
from skimage import io
import matplotlib.pyplot as plt
import math

def posterize(image):
    rows,cols,dims  = image.shape
    for i in range(0,rows):
           for j in range(0,cols):
                lun = (image[i,j,0]+image[i,j,0]+image[i,j,2])/3
                
                if(lun<32):
                    image[i,j,0] = 0
                    image[i,j,0] = 0
                    image[i,j,0] = 0
                else:
                    image[i,j,0] = 255
                    image[i,j,1] = 255
                    image[i,j,2] = 255
    return image
    
img =io.imread('Santa Claus.PNG')
io.imshow(img)

img2 = posterize(img)
plt.figure()
io.imshow(img2)
io.show()
```

```python
from skimage import io
import matplotlib.pyplot as plt
import numpy as np

def reduce(image):
    rows,cols,dims  = image.shape
    retimg = np.uint8(np.zeros((rows,cols,3)))
    for i in range(0, rows):
        for j in range(0,cols):
            lum = 0
            for k in range(0,3):
                lum = lum + image[i,j,k]
            lum = lum / 3
            if lum <64:
                retimg[i,j,0] = 0
                retimg[i,j,1] = 0
                retimg[i,j,2] = 0
            elif lum > 120:
                retimg[i,j,0] = 255
                retimg[i,j,1] = 255
                retimg[i,j,2] = 255
            else:
                retimg[i,j,0] = 255
                retimg[i,j,1] = 255
                retimg[i,j,2] = 0
                
    return  retimg

img = io.imread('Santa Claus.PNG')
io.imshow(img)

img2 = reduce(img)
plt.figure()
io.imshow(img2)
io.show()
```

```python
from skimage import io
import matplotlib.pyplot as plt
import math

def sep(image):
    rows,cols,dims  = image.shape
    for i in range(0,rows):
           for j in range(0,cols):
                red = image[i,j,0]
                blue = image[i,j,2]
                
                if (red<63):
                    red =  red*1.1
                    blue = blue *0.9
                    
                if(red >62 and red <192):
                    red =  red *1.15
                    blue = blue *0.85
                    
                if(red >191):
                    red = red*1.08
                if (red >255):
                    red = 255
                
                blue =  blue *0.93
                
                image[i,j, 0] = red
                image[i,j,1] = blue
                
    return image


img = io.imread('black_man.jpg')
io.imshow(img)

img2 = sep(img)
plt.figure()
io.imshow(img2)
io.show()
```

```python
from skimage import io
import matplotlib.pyplot as plt
import math

def sep(image):
    rows,cols,dims  = image.shape
    for i in range(0,rows):
           for j in range(0,cols):
                red = image[i,j,0]
                blue = image[i,j,2]
                
                if (red<63):
                    red =  red*1.1
                    blue = blue *0.9
                    
                if(red >62 and red <192):
                    red =  red *1.15
                    blue = blue *0.85
                    
                if(red >191):
                    red = red*1.08
                if (red >255):
                    red = 255
                
                blue =  blue *0.93
                
                image[i,j, 0] = red
                image[i,j,1] = blue
                
    return image


img = io.imread('road.jpg')
plt.figure(num = 'road',figsize = (15,15))

plt.subplot( 2,2,1)
plt.title('origin image')
plt.imshow(img)

img2 = sep(img.copy())
plt.subplot( 2,2,2)
plt.title('sepia toned1')
plt.imshow(img2)

img3 = sep(img2.copy())
plt.subplot( 2,2,3)
plt.title('sepia tonsd2')
plt.imshow(img3)

img4 = sep(img3.copy())
plt.subplot( 2,2,4)
plt.title('sepia tones3')
plt.imshow(img4)
plt.show()
```

```python
from skimage import io
import matplotlib.pyplot as plt


def edg(image):
    rows,cols,dims  = image.shape
    for i in range(0,rows-1):
           for j in range(0,cols-1):
                sum = 0
                sum_right =0
                for k in range(0,3):
                    sum = sum + image[i,j,k]
                    sum_right = sum_right + image[i+1,j+1,k]
                dif = abs(sum_right - sum)
                
                for k in range(0,3):
                    image[i,j,k] = dif
    return image

img = io.imread('road.jpg')
plt.figure(num = 'road',figsize = (20,20))

plt.subplot(1,2,1)
plt.title('origin image')
plt.imshow(img)

img2 = edg(img)
plt.subplot(1,2,2)
plt.title('edge detected imge')
plt.imshow(img2)
plt.show()
```

```python
from skimage import io
import matplotlib.pyplot as plt


def edg(image,thre):
    rows,cols,dims  = image.shape
    for i in range(0,rows-1):
           for j in range(0,cols-1):
                sum = 0
                sum_right =0
                for k in range(0,3):
                    sum = sum + image[i,j,k]
                    sum_right = sum_right + image[i+1,j+1,k]
                dif = abs(sum_right - sum)
                if dif >  thre:
                    value = 0
                else:
                    value = 255
                for k in range (0,3):
                    image[i,j,k] = value
                    
    return image

img = io.imread('road.jpg')
threshold = 50

plt.figure(num = 'road',figsize = (20,20))

plt.subplot(1,2,1)
plt.title('origin image')
plt.imshow(img)

img2 = edg(img,threshold)
plt.subplot(1,2,2)
plt.title('edge detected imge')
plt.imshow(img2)
plt.show()
```
