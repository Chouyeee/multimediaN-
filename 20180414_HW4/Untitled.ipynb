```python
class Man: 
    def __init__(self,name):
        self.name = name
        print("Initialized")
        
    def hello(self):
        print("Hello" + self.name+ "!")
        
    def goodbye(self):
        print("Good-Bye"+self.name+"!")
```

```python
m= Man("Tom")
m.hello()
m.goodbye()
```

```python
def factorial(g):
    if g>0:
        return g*factorial(g-1)
    else:
        return 1
```

```python
factorial(5)
```

```python
def fib(n):
    a ,b = 0,1
    while a<n:
         print(a,end = '')
         a,b=b,a+b
    print()
```

```python
fib(2000)
```
